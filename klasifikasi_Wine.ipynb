```python
import pandas as pd
wine = pd.read_csv('/content/wine_dataset.csv')
```

```python
wine.info()
wine.head(10)
```

```python
#memisah atribut n rabel
X = wine[['fixed_acidity','volatile_acidity','citric_acid','residual_sugar','chlorides','free_sulfur_dioxide','total_sulfur_dioxide','density','pH','sulphates','alcohol','quality']]
y = wine['style']
#bagi data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
#model dt
tree_model = DecisionTreeClassifier()
#latih model menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)
```

```python
#evaluasi mdoel
from sklearn.metrics import accuracy_score
y_pred = tree_model.predict(X_test)
acc_score = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy:', acc_score)
```

```python
#prediksi model dengan tree_model (['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
print(tree_model.predict([[7.3, 0.22, 0.37, 15.5, 0.048, 70, 203, 0.99899, 3.25, 0.77, 9.4, 5]])[0])
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file = "wine_tree.dot",
    feature_names = ['fixed_acidity','volatile_acidity','citric_acid','residual_sugar','chlorides','free_sulfur_dioxide','total_sulfur_dioxide','density','pH','sulphates','alcohol','quality'],
    class_names = ['red', 'white'],
    rounded = True,
    filled = True
)
```
